```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming tips is your DataFrame
sns.displot(tips['total_bill'], kde=False)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming tips is your DataFrame
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.title('Joint Plot of Total Bill vs. Tip')  # This adds a title to the figure
plt.show()
```

```python
# Assuming iris is your DataFrame
g = sns.PairGrid(iris)
g.map_diag(plt.hist)      # Diagonal: Histograms
  # Lower: Kernel density estimate plots

plt.show()
```
