# Which is the most and least cost depreciated vehicle in data ?

```python
data['Depreciation'] = data['Selling_Price'] - data['Present_Price']
most_depreciated_vehicle = data.loc[data['Depreciation'].idxmax()]

least_depreciated_vehicle = data.loc[data['Depreciation'].idxmin()]

print("Most depreciated vehicle:")
print(most_depreciated_vehicle[['Car_Name', 'Depreciation']])

print("\nLeast depreciated vehicle:")
print(least_depreciated_vehicle[['Car_Name', 'Depreciation']])
```
